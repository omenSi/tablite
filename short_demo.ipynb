```python
import tempfile, pathlib
d = tempfile.gettempdir()
path = pathlib.Path(d) / "large.csv"
```

```python
with path.open('r') as fi:
    print(fi.read(1000))
```

```python
from tablite import Table
```

```python
Table.reset_storage()
```

NB- old dce takes ~30 minutes.

```python
my_table = Table.import_file(path, import_as='csv')
```

```python
my_table.show()
```

```python
mytable2 = Table.import_file(path, import_as='csv')  
mytable2.show()
```

```python
copy_of_mytable = my_table.copy()
copy_of_mytable.show()
```

```python
a_bigger_table = mytable2 + copy_of_mytable + my_table
a_bigger_table.show()
```

```python
random_slice = my_table[slice(3,200,18)]
random_slice.show()
```

```python
from tablite import GroupBy 
product_category, temp_zone = '6', '8'

pivot = my_table.pivot(rows=[product_category], columns=[temp_zone], functions=[('6', GroupBy.count)])
```

```python
pivot.show()
```

```python
datum, volume = '2','5'
grpby = my_table.groupby(keys=[datum], functions=[(volume, GroupBy.max),(volume, GroupBy.min)])
grpby.show()
```

```python
sorted_grpby = grpby.sort(**{datum:False})
sorted_grpby.show(slice(0,None))
```

```python
import matplotlib.pyplot as plt
plt.plot(sorted_grpby['2'], sorted_grpby['Max(5)'])
plt.title("Dovydas charts are better than Bjorns")
plt.show()
```
